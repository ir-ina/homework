# 13. Выведите количество игр которая сыграла каждая команда в данном сезоне?

```python
home_games = data['home_team'].value_counts()
away_games = data['away_team'].value_counts()
total_games = pd.concat([home_games, away_games], axis = 1 )
total_games['total'] = total_games.home_team+total_games.away_team
total_games.head(10)
```

# 14. Какая команда выиграла наибольшое количество матчей в сезоне? (Будьте внимательны с типом матча в сезоне - season).

```python
df_home_wins = pd.pivot_table (data, index = 'home_team', values = 'home_team_win', aggfunc= np.sum)
df_away_wins = pd.pivot_table (data, index = 'away_team', values = 'home_team_loss', aggfunc=np.sum)
df_wins = pd.concat([df_home_wins, df_away_wins], axis = 1 )
df_wins ['total_wins'] = df_wins.home_team_win+df_wins.home_team_loss
df_wins.sort_values( by = 'total_wins', ascending = False ).iloc[0]
```

# 15. Какая команда выиграла наибольшее количество домашних матчей в сезоне?

```python
home_winner = data.groupby(['home_team'],as_index=False).aggregate({'home_team_win': "sum"}).sort_values("home_team_win", ascending = False)
home_winner.iloc[[0]]
```

# 16. Какая команда выиграла наибольшее количество гостевых матчей в сезоне?

```python
away_winner = data.groupby(['away_team'],as_index=False).aggregate({'home_team_loss': "sum"}).sort_values("home_team_loss", ascending = False)
away_winner.iloc[[0]]
```

# 17. Какая команда проиграла наибольшее количество матчей в сезоне?

```python
df_home_losses = pd.pivot_table (data, index = 'home_team', values = 'home_team_loss', aggfunc= np.sum)
df_away_losses = pd.pivot_table (data, index = 'away_team', values = 'home_team_win', aggfunc=np.sum)
df_losses = pd.concat([df_home_losses, df_away_losses], axis = 1 )
df_losses ['total_losses'] = df_losses.home_team_win+df_losses.home_team_loss
df_losses.sort_values( by = 'total_losses', ascending = False ).iloc[0]
```

# 19. Правда ли что большинство проигрышных домашних матчей приходятся на Субботу и Воскресенье?

```python
def if_weekend(row):
    """
    Функция отнесения дней недель к выходным/будним
    """
    if row['day_of_week'] == ('Saturday' or 'Sunday'):
        return 'weekend'
    else:
        return 'weekday'
    
data['weekend_weekday'] = data.apply( if_weekend, axis = 1 )
data.groupby(['weekend_weekday'],as_index=False).aggregate({'home_team_loss': "sum"})
```

# 20. Правда ли что наибольшее количество ранов происходит в холодную погоду? (Холодной погодой считается погода ниже 0 градусов)

```python
data['temperature_celsius'] = (data['temperature'].apply(lambda x:( x - 32)/1.8))
def if_cold(row):
    """
    Функция отнесения дней к холодным/теплым
    """
    if row['temperature_celsius'] < 0:
        return 'cold'
    else:
        return 'warm'

data['cold_warm'] = data.apply( if_cold, axis = 1 )
data.groupby(['cold_warm'], as_index = False).aggregate({'total_runs': "sum"})
```

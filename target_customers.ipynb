```python
import pandas as pd
import numpy as np

from plotly.offline import init_notebook_mode, iplot
import plotly
import plotly.graph_objs as go

init_notebook_mode(connected=True)
```

```python
data=pd.read_csv(r'G:\__ПРЕЗЕНТАЦИИ\descriptive_stats\clients.csv')
data.head()
```

```python
data.info()
```

```python
# Так как нас интересует портрет человека, уберем данные по компаниям

data = data.drop(data[data.Entity == 'Firm'].index)
data.info()
```

```python
# Убираем знак $ и запятые из цены

def clearPrice(row):
    row['Price'] = row['Price'][3:]
    row['Price'] = row['Price'].replace(',', '')
    return row['Price']
    
data['Price'] = data.apply( clearPrice, axis = 1 )
data.head()
```

```python
# Поменяем типы данных цены и возраста

data['Price'] = data.Price.astype('float64')
data['Age at time of purchase'] = data['Age at time of purchase'].astype('int64')
data.info()
```

```python
# Переведем площадь в м2

data['Area (m)'] = (data['Area (ft.)'].apply(lambda x: x/10.764))
data.head()
```

```python
# Найдем типичный пол

data['Gender'].value_counts()
```

```python
# Найдем типичную возрастную группу
df_ages = data.groupby('Age Interval ').aggregate({'Age Interval ': "count"})
df_ages.columns = ['Amount']
df_ages
```

```python
trace0 = go.Scatter(
    x=df_ages.index,
    y=df_ages.Amount,
    name='Age Interval ')

data = [trace0]
layout = {'title': 'Statistics of age at purchase'}

fig = go.Figure(data=data, layout=layout)

iplot(fig, show_link=False)
```

```python
# Найдем типичный тип недвижимости

data['Type of property'].value_counts()
```

```python
# Найдем типичную цель покупки

data['Purpose'].value_counts()
```

```python
# Найдем типичную площадь недвижимости

data['Area (m)'].median()
```

```python
# Найдем типичный чек

data['Price'].median()
```

```python
# Найдем типичный источник лида

df_source = data.groupby('Source').aggregate({'Source': "count"})
df_source.columns = ['Amount']
df_source
```

```python
labels = df_source.index
values = df_source.Amount

fig = go.Figure(data=[go.Pie(labels=labels, values=values)])
fig.show()
```
